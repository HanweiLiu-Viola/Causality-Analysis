# Start Here: Environment and Container

This notebook is the entry point for the Section 7 tutorial. It checks the host tools needed to run the project, then shows how to start the pinned Docker image and open JupyterLab with this repository mounted.

**Input:** a local clone of this repository.

**Output:** a verified host environment and a running containerized Jupyter session.

**Tutorial step:** Section 7.2, run the reproducible container.

---

## 1. Check Required Tools

Run this cell on the host machine before starting the container.


````python
import subprocess
import shutil

checks = {
    "git":            ["git", "--version"],
    "conda":          ["conda", "--version"],
    "docker":         ["docker", "--version"],
    "snakemake":      ["snakemake", "--version"],
    "node":           ["node", "--version"],
    "bids-validator": ["bids-validator", "--version"],
}

print(f"{'Tool':<20} {'Status':<10} {'Version'}")
print("-" * 55)
for name, cmd in checks.items():
    if shutil.which(cmd[0]) is None:
        print(f"{name:<20} {'NOT FOUND':<10}")
    else:
        result = subprocess.run(cmd, capture_output=True, text=True)
        version = (result.stdout or result.stderr).strip().split("\n")[0]
        print(f"{name:<20} {'OK':<10} {version}")
````

All six tools should show **OK**. If any show **NOT FOUND**, install that tool on the host machine before continuing.

---

## 2. Start the Container

Pull and run the pinned tutorial image from the repository root:

```bash
docker pull viola1003/causality-analysis:v1.1.0
docker run -p 8888:8888 -v "$(pwd):/home/jovyan/work" viola1003/causality-analysis:v1.1.0
```

Open the JupyterLab URL printed by Docker, then continue with the next notebook inside the mounted `work/` directory.
